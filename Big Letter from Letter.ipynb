```python
def advancedPrint(char, word, axis=0):
    if len(char) != 1:
        print('The character used as a component must be only one character.')
    
    row = 5
    alpha = list(''.join('ABCDEFGHIJKLMNOPQRSTUVWXYZ'.split()))

# To print horizontally

    if axis == 0:
        for i in range(row):
            lst_print = [char_struc[alpha.index(n.upper())][i][0].replace('1', char) if n != ' ' else '      ' for n in word]
            print(*lst_print)
            
# To print vertically
    elif axis == 1:
        for i in range(len(word)):
            if word[i] != ' ':
                ind = alpha.index(word[i].upper())
                for j in range(row):
                    print(char_struc[ind][j][0].replace('1', char))
            else:
                print('\n' * 2)
            print('\n')
            
    else:
        print('Please iput the correct axis. (0 = x-axis, 1 = y-axis)')
```
